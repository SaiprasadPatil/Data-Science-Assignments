Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sms
import plotly.express as px
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
tips_data = px.data.tips()
box_plot = px.box(tips_data, x='day', y='total_bill', title='Box Plot of Total Bill Amounts by Day')
box_plot.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
histogram = px.histogram(
    tips_data,
    x='sex',
    y='total_bill',
    color='day',
    pattern_shape='smoker',
    title='Histogram of Total Bill Amounts by Sex',
)

histogram.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
iris_data = px.data.iris()
scatter_matrix = px.scatter_matrix(
    iris_data,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset",
)

scatter_matrix.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

A distplot (distribution plot) is a type of plot that displays the distribution of a continuous variable. It combines a histogram with a kernel density estimate, providing insights into the shape, spread, and central tendency of the data.

Using Plotly Express, you can create a distplot with the px.histogram function and the histfunc='density' parameter to include the kernel density estimate. Here's an example:

```python
import plotly.express as px
import seaborn as sns
import numpy as np

tips_data = px.data.tips()

histogram = px.histogram(
    tips_data,
    x="total_bill",
    color="sex",
    marginal="rug",  
    title="Distribution Plot of Total Bill Amounts by Sex",
)

kde_values = sns.kdeplot(tips_data["total_bill"], fill=False).get_lines()[0].get_ydata()

kde_scatter = px.line(
    x=tips_data["total_bill"].unique(),
    y=kde_values,
    mode="lines",
    name="KDE",
)

histogram.add_trace(kde_scatter.data[0])

histogram.show()
```

Creating a combined histogram and KDE-like visualization in Plotly Express might be challenging due to the lack of direct support for KDE. As of my last knowledge update in September 2021, Plotly Express primarily focuses on creating individual plots rather than combining a histogram and a KDE plot.
