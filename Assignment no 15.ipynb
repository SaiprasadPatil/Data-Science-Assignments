Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for Python. It is designed to be simple, lightweight, and flexible, allowing developers to quickly build web applications with minimal code and effort. Flask follows the WSGI (Web Server Gateway Interface) standard and is based on the Werkzeug WSGI toolkit and the Jinja2 template engine.

Advantages of Flask Framework:
Lightweight and Minimalistic: Flask has a minimalistic core with a small set of dependencies.
Easy to Get Started: Flask is straightforward to get started with, making it a great choice for beginners. 
Flexibility: Flask gives developers the freedom to choose the tools and libraries they prefer. 
Modular and Extensible: Flask follows a modular design, allowing developers to add or remove components as needed.
Powerful Routing System: Flask provides a flexible and powerful routing system that allows developers to map URLs to specific view functions easily.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs (or routes) to specific functions in your Flask application. It allows you to define how your application responds to different URLs by associating each URL with a corresponding view function.

In Flask, the @app.route() decorator is used to define routes. This decorator binds a URL pattern to a view function, which is executed when a request is made to that URL. The view function generates the response that is sent back to the client.

App routes are used in Flask to define the different pages and functionality of a web application. They allow you to structure and organize your application's functionality by associating specific URLs with the corresponding actions or content. App routes make it easy to handle different HTTP methods (such as GET or POST) and provide dynamic behavior by using parameters in the URL.

Using app routes helps in building a clean and logical URL structure for your application, making it easier for users to navigate and understand the functionality of different pages. It also enables the separation of concerns by assigning specific tasks or functionality to different view functions.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used for URL building. It is a built-in function provided by Flask that generates a URL for a given view function.

```python
from flask import Flask, redirect,url_for
 
   
app = Flask(__name__)  
  
@app.route('/ios')
def ios():
    return '<h1>User Selected IOS<h1>'

@app.route('/windows')
def windows():
    return '<h1>User Selected Windows<h1>'

@app.route('/android')
def android():
    return '<h1>User Selected Android<h1>'

@app.route('/prod/<name>')
def user(name):
    if name == 'ios':
        return redirect(url_for('ios'))
    if name == 'windows':
        return redirect(url_for('windows'))
    if name == 'android':
        return redirect(url_for('android'))

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
