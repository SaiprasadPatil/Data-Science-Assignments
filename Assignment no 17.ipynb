Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

On pandas, a DataFrame is a 2-dimensional labeled data structure, similar to a spreadsheet or SQL table. It consists of rows and columns, where each column can hold data of different types (e.g., integers, floats, strings) and is given a unique label. DataFrame is one of the most commonly used data structures in pandas and is suitable for representing and working with structured, tabular data.

On the other hand, a pandas Series is a 1-dimensional labeled array that can hold data of a single data type. It is similar to a Python list or a NumPy array but has additional features such as labeled indexing, which makes it more versatile for data manipulation and analysis.

Here's an example to illustrate the difference between a DataFrame and a Series:

```python
b=pd.Series(["Sai","Patil","SaiPatil"])
```

```python
b
```

```python
c=pd.DataFrame(["Sai","Patil","SaiPatil"])
```

```python
c
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

Among the options provided (Series, DataFrame, Panel), the mutable nature refers to the ability of a data structure to be modified after its creation. In pandas, the Series and DataFrame are mutable, while the Panel is not.

Series: Mutable,
DataFrame: Mutable,
Panel: Not mutable (Deprecated).

It's worth noting that the Panel data structure has been deprecated and users are encouraged to use multi-dimensional arrays or DataFrames for handling higher-dimensional data. As a result, you are unlikely to encounter Panels in recent versions of pandas. For multi-dimensional data, you can use DataFrames, which are versatile and more commonly used in practice
