Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating a variety of informative and attractive statistical graphics. Here are five types of plots that you can create using Seaborn, along with their common uses:

1. **Histogram:**
   - Use: Histograms are used to visualize the distribution of a single variable. They show the frequency or count of data points falling within specified intervals (bins).
   - Seaborn Function: `seaborn.histplot()`

2. **Bar Plot:**
   - Use: Bar plots are used to compare the values of different categories or groups. They are suitable for showing categorical data and comparing quantities across different groups.
   - Seaborn Function: `seaborn.barplot()`

3. **Heatmap:**
   - Use: Heatmaps are used to display the relationships between two categorical variables in a matrix format. They are often used to visualize correlations or associations in a dataset.
   - Seaborn Function: `seaborn.heatmap()`

4. **Pair Plot:**
   - Use: Pair plots are used to visualize the relationships between multiple variables in a dataset. They create scatter plots for all combinations of variables, making it easy to identify patterns and correlations.
   - Seaborn Function: `seaborn.pairplot()`

5. **Box Plot:**
   - Use: Box plots (box-and-whisker plots) are used to summarize the distribution of a dataset and identify potential outliers. They display the median, quartiles, and potential outliers of numerical data.
   - Seaborn Function: `seaborn.boxplot()`

These are just a few examples of the many types of plots you can create using Seaborn. Seaborn also provides functions for creating violin plots, swarm plots, count plots, point plots, and more. Each of these plot types has its own unique use case and can be highly informative for visualizing different aspects of your data.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
fmri = sns.load_dataset("fmri")
plt.figure(figsize=(10, 6))  # Optional: Adjust figure size
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset("titanic")
plt.figure(figsize=(12, 6)) 
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Age vs. Pclass")
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Fare vs. Pclass")

plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
plt.figure(figsize=(10, 6)) 
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title("Price Distribution by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend(title="Cut")
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue='species')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
plt.figure(figsize=(10, 6))  # Optional: Adjust figure size
sns.heatmap(data=flights_pivot, cmap='coolwarm', annot=True, fmt='d')
plt.title("Flights Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
